Q25. Nested if: If a number is positive, check if it's prime or not.

```python
num = int(input("Enter a number: "))

if num > 0:
    if num == 1:
        print("1 is neither prime nor composite")
    else:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print("Prime number")
        else:
            print("Not a prime number")
else:
    print("Number is not positive")
```
